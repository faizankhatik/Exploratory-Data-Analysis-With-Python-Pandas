# Combine all filters together

```python
df2 = df[(df['Event distance/length'].isin(['50mi','50km'])) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

# Remove (USA) from event name

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

# Clean-up Athlete Age

```python
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

# Remove h from athlete performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

# Drop Columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age category

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete age'].isna()==1]
```

```python
df2 = df2.dropna()
```

# Reset Index

```python
df2.reset_index(drop=True)
```

# Fix Types

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
import pandas as pd

# Assuming df2 is your DataFrame

# Check if the 'Athlete average speed' column is not of object (string) type
if df2['Athlete average speed'].dtype != object:
    # If not string type, convert it to string first
    df2['Athlete average speed'] = df2['Athlete average speed'].astype(str)

# Now you can safely use .str accessor
df2['Athlete average speed'] = df2['Athlete average speed'].str.extract(r'(\d+\.?\d*)').astype(float)

# Convert to timedelta, assuming the unit is km/h
# We will assume that the average speed is in km/h and convert to seconds
df2['Athlete average speed'] = pd.to_timedelta(df2['Athlete average speed'], unit='h')

# Extract total seconds and convert to float
df2['Athlete average speed'] = df2['Athlete average speed'].dt.total_seconds().astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

# Rename Columns


```python
'''

Year of event	int64
Event dates	object
Event name	object
Event distance/length	object
Event number of finishers	float64
Athlete performance	object
Athlete gender	object
Athlete average speed	float64
Athlete ID	float64
Athlete age	int64

'''
```

```python
df2 = df2.rename(columns={'Year of event':'year','Event dates':'race_day','Event name':'race_name','Event distance/length':'race_length','Event number of finishers':'race_number_of_finishers','Athlete performance':'athlete_performance','Athlete gender':'athlete_gender','Athlete average speed':'athlete_average_speed','Athlete ID':'athlete_id'})
df2 = df2.rename(columns={'Athlete age':'athlete_age'})
```

```python
df2.head()
```

# Re-order Columns

```python
df3 = df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed','year']]
```

```python
df3.head()
```

# Season for data -> slower in summer than winter

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']) .sort_values('mean', ascending=False)
```
